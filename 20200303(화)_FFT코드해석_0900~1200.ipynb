```python
#!/usr/bin/python
#-*- coding:cp949 -*-
# 1~2   .
"""
FFT Test code in python

Withrobot Lab.
http://Withrobot.com
2008.4.28
Original link: https://withrobot.tistory.com/217
"""

#pylab matplotlib     
#= matplotlib.pyplot numpy     #:        
                                                                         # import:  
                                                        # numpy: array         
                                                        # array:    . ()
#fft, fftshift:  
from pylab import* 
from numpy.fft import fft, fftshift

Fs = 1000.     # the sampling frequency
Ts = 1./Fs     # the sampling peroid

N = 256        #  
freqStep = Fs/N # resolution of the frequency in frequency domain:      (?)

f = 10*freqStep    # frequency of the wave
t = arange(N)*Ts   # x ticks(, ) in time domain, t = N*Ts : 
y = cos(2*pi*f*t) + 0.5*sin(2*pi*3*f*t)  #  

Y = fft(y)    # FFT 
Y = fftshift(Y) # middles the zero-point's axis

figure(figsize=(8,8))
subplots_adjust(hspace=.4) # hspace, wspace  subplot     subplot   
                                                                     # subplot(m, n, p)   mxn  , p 
                                                                     #     .

# Plot time data
subplot(3,1,1)
plot(t, y, '.-')
grid("on")
xlabel('Time (seconds)')
ylabel('Amplitude')
title('signals')
axis('tight')

freq = freqStep * arange(-N/2, N/2)  # x ticks in frequency domain

# Plot spectral magnitude
subplot(3,1,2) #  plot       subplot  
plot(freq, abs(Y), '.-b') # plot()   plt.plot(x, y, linewidth=2.0)   .
grid("on")
xlabel('Frequency')
ylabel('Magnitude (Linear)')

# Plot phase
subplot(3,1,3)
plot(freq, angle(Y), '.-b')
grid("on")
xlabel('Frequency')
ylabel('Phase (Radian)')

show()
```
