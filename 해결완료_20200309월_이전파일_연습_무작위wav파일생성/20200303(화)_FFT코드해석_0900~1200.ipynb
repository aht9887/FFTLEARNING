```python
#!/usr/bin/python
#-*- coding:cp949 -*-
#상기의 1~2 줄은 없어도 무방하다.
"""
FFT Test code in python

Withrobot Lab.
http://Withrobot.com
2008.4.28
Original link: https://withrobot.tistory.com/217
"""

#pylab은 matplotlib을 설치할 때 같이 설치되는 모듈
#=대부분의 matplotlib.pyplot과 numpy를 하나의 네임스페이스에 임폴트 시킴 #이름공간: 특정한 객체를 이름에 따라 구분할 수 있는 범위
                                                                         # import: 모듈 가져오기
                                                        # numpy: array 단위로 데이터를 관리하여 벡터 및 행렬 계산을 하는 라이브러리
                                                        # array: 여러 데이터를 담기위한 자료구조. (배열)
#fft, fftshift: 푸리에 변환
from pylab import* 
from numpy.fft import fft, fftshift

Fs = 1000.     # the sampling frequency
Ts = 1./Fs     # the sampling peroid

N = 256        # 샘플 개수
freqStep = Fs/N # resolution of the frequency in frequency domain: 주파수를 독립변수로 하는 영역에서의 주파수 해결(해석?)

f = 10*freqStep    # frequency of the wave
t = arange(N)*Ts   # x ticks(점, 순간) in time domain, t = N*Ts : 좌표설정
y = cos(2*pi*f*t) + 0.5*sin(2*pi*3*f*t)  # 테스트 신호

Y = fft(y)    # FFT 분석
Y = fftshift(Y) # middles the zero-point's axis

figure(figsize=(8,8))
subplots_adjust(hspace=.4) # hspace, wspace의 경우 subplot이 여러 개 일 때 subplot 간의 간격을 의미함
                                                                     # subplot(m, n, p)는 현재 그림을 mxn 그리드로 나누고, p로 
                                                                     # 로 지정된 위치에 좌표축을 만든다.

# Plot time data
subplot(3,1,1)
plot(t, y, '.-')
grid("on")
xlabel('Time (seconds)')
ylabel('Amplitude')
title('signals')
axis('tight')

freq = freqStep * arange(-N/2, N/2)  # x ticks in frequency domain

# Plot spectral magnitude
subplot(3,1,2) # 한 plot에 두 그림을 나란히 놓고 비교하기 위해 subplot을 사용하는 것
plot(freq, abs(Y), '.-b') # plot() 그래프 그리기 plt.plot(x, y, linewidth=2.0)과 같이 이해하면 될듯하다.
grid("on")
xlabel('Frequency')
ylabel('Magnitude (Linear)')

# Plot phase
subplot(3,1,3)
plot(freq, angle(Y), '.-b')
grid("on")
xlabel('Frequency')
ylabel('Phase (Radian)')

show()
```
